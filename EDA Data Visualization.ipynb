### The `plt.scatter()` function is a part of the matplotlib library in Python, which is used to plot a scatter plot. Scatter plots’ primary uses are to observe and show relationships between two numeric variables.

#### In your code, `plt.scatter(data['day'], data['tip'], c=data['size'], s=data['total_bill'])`, here's what each parameter does:

#### - `data['day']` and `data['tip']`: These are the x and y coordinates of each point in the scatter plot. The points are placed on the graph corresponding to their 'day' and 'tip' values.

#### - `c=data['size']`: This parameter is for color. It means that the color of points will be determined by the 'size' column of your data. If 'size' is numeric, the colors will be chosen from a color scale. If 'size' is categorical (like strings or factors), different colors will be used for each category.

#### - `s=data['total_bill']`: This parameter is for size. It means that the size of points will be determined by the 'total_bill' column of your data. Larger values will create larger points.

#### So, this line of code will create a scatter plot where each point represents a row in your data. The position of the point corresponds to its 'day' and 'tip' values, the color corresponds to its 'size' value, and the size of the point corresponds to its 'total_bill' value.

### In Python’s matplotlib library, autopct='%1.2f%%' is a format string used in pie charts to label the wedges with their numeric value. The autopct parameter allows you to display the percent value using Python string formatting. Here is what it means:

#### % is a placeholder for where the numbers will be substituted.
#### 1.2f specifies the format of the number. 1 is the minimum width of the field, 2 is the number of digits after the decimal point, and f stands for float.
#### The final %% is how you can display an actual % sign in a format string (since % is usually used for other purposes), so this gets displayed as a single % on the chart.
#### If you want to display the frequency (or count) instead of the percentage, you can define your own autopct function. Here’s an example:

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
#data = pd.read_csv('tips.csv')

# initializing the data
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]

# plotting the data
plt.pie(data, labels=cars, autopct='%1.2f%%')

# Adding title to the plot
plt.title("Car data")

plt.show()
```

The linewidth parameter is used to specify the width of the edges. You’ve set linewidth=2, which means the edges will have a width of 2 units. So, in your plot, each bar will have a green color fill with blue edges that are 2 units wide.

```python
import matplotlib.pyplot as plt

# initializing the data
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR']
data = [23, 13, 35, 15, 12]

explode = [0.1, 0.5, 0, 0, 0]
colors = ("orange", "cyan", "yellow", "grey", "green")

# This function will be called for each slice in the pie chart. It returns a string to be displayed on the slice.
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% ({v:d})'.format(p=pct,v=val)
    return my_autopct

# plotting the data
plt.pie(data, labels=cars, explode=explode, autopct=make_autopct(data), colors=colors, shadow=True)

plt.show()
```

The function `make_autopct(values)` is a higher-order function, which means it's a function that returns another function. In this case, it returns the function `my_autopct(pct)`.

Here's a breakdown of what each part does:

- `def make_autopct(values):` This is the definition of the outer function `make_autopct`. It takes one argument, `values`, which is a list of the data values that are being plotted in the pie chart.

- `def my_autopct(pct):` This is the definition of the inner function `my_autopct`. It takes one argument, `pct`, which represents a percentage of the total.

- `total = sum(values)` This line calculates the total sum of all data values.

- `val = int(round(pct*total/100.0))` This line calculates the actual data value that corresponds to the percentage `pct`. It does this by multiplying the total by the percentage (divided by 100 to convert it from a percentage to a proportion), rounding the result to the nearest whole number, and converting that number to an integer.

- `return '{p:.2f}% ({v:d})'.format(p=pct,v=val)` This line returns a string that contains both the percentage (formatted to two decimal places) and the actual data value. The `{p:.2f}` part is a placeholder for the percentage, and `{v:d}` is a placeholder for the data value. The `.format(p=pct,v=val)` part replaces these placeholders with the actual values of `pct` and `val`.

- `return my_autopct` This is the last line of the outer function `make_autopct`. It returns the inner function `my_autopct`.

So when you call `make_autopct(data)`, it returns a new function that takes a percentage as an argument and returns a string that represents both that percentage and the corresponding data value. This returned function can then be used as an argument for `autopct` in `plt.pie()`, which determines what text to display on each slice of the pie chart.

```python
import matplotlib.pyplot as plt

# Initializing the data
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR']
data = [23, 10, 35, 15, 12]

# Plotting the data
plt.pie(data, labels=cars, autopct=lambda p : '{:.0f}'.format(p*sum(data)/100))

# Adding title to the plot
plt.title("Car data")

plt.show()
```

- `plt.pie(data, labels=cars, autopct=lambda p : '{:.0f}'.format(p*sum(data)/100))`

### This line is using the `pie` function from the `matplotlib.pyplot` module (which is imported as `plt`) to create a pie chart.

##### Here's what each part does:

##### - `data`: This is the array of data that you want to plot. Each number in the array corresponds to a slice of the pie. The size of each slice is proportional to the corresponding data value.

##### - `labels=cars`: This assigns labels to each slice of the pie. The `cars` array contains the labels.

##### - `autopct=lambda p : '{:.0f}'.format(p*sum(data)/100)`: This is a bit more complex. It's using a lambda function to format the labels for each slice of the pie. Let's break it down:
#####  - `autopct` is a parameter that accepts a string format or a function. If it's a string format (like '%1.1f%%'), it'll display the percentage value on each slice. But here, we're passing a function to display actual values (frequency) instead of percentages.
#####  - `lambda p : '{:.0f}'.format(p*sum(data)/100)` is a lambda function. It's an anonymous, inline function that's defined right where we use it.
#####  - `p` is the parameter to this function. When the pie chart is being drawn, matplotlib will call this function once for each slice, passing in the percentage value of the slice.
#####  - `'{:.0f}'.format(p*sum(data)/100)` takes that percentage, multiplies it by the total of all data (to get back to the original value), and then formats it as an integer (because of the `:0f` in the format string). This value is then displayed on each slice of the pie.



## A histogram is basically used to represent data in the form of some groups. 
#### It is a type of bar plot where the X-axis represents the bin ranges while the Y-axis gives information about frequency. 
#### The hist() function is used to compute and create a histogram. In histogram, if we pass categorical data then it will automatically compute the frequency of that data i.e. how often each value occurred.

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("C:/Users/faraz211724/Downloads/tips.csv")

# histogram of total_bills
plt.hist(data['total_bill'])

plt.title("Histogram")

# Adding the legends
plt.show()
```

### Customization that is available for the Histogram – 

#### bins: Number of equal-width bins 
#### color: For changing the face color
#### edgecolor: Color of the edges
#### linestyle: For the edgelines
#### alpha: blending value, between 0 (transparent) and 1 (opaque)

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading the tips.csv file
data = pd.read_csv('C:/Users/faraz211724/Downloads/tips.csv')

# initializing the data
x = data['total_bill']

# plotting the data
plt.hist(x, bins=25, color='green', edgecolor='blue',
		linestyle='--', alpha=0.5)

# Adding title to the plot
plt.title("Tips Dataset")

# Adding label on the y-axis
plt.ylabel('Frequency')

# Adding label on the x-axis
plt.xlabel('Total Bill')

plt.show()
```

bins: This parameter specifies the number of bins or the specific bin edges to be used in the histogram. It determines the number of intervals into which the data is divided and displayed as bars in the histogram. In this case, bins=25 means that the histogram will have 25 bins or bars.

alpha: This parameter controls the transparency of the bars in the histogram. It is a value between 0 and 1, where 0 means completely transparent (invisible) and 1 means completely opaque. In this case, alpha=0.5 means that the bars will be semi-transparent, allowing some visibility of the underlying plot or data points.

### Using subplots_adjust() method to set the spacing between subplots
#### We can use the plt.subplots_adjust() method to change the space between Matplotlib subplots. The parameters wspace and hspace specify the space reserved between Matplotlib subplots. They are the fractions of axis width and height, respectively. And the parameters left, right, top and bottom parameters specify four sides of the subplots’ positions. They are the fractions of the width and height of the figure.

```python
# importing packages
import numpy as np
import matplotlib.pyplot as plt

# create data
x=np.array([1, 2, 3, 4, 5])

# making subplots
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x*x)
ax[1, 1].plot(x, x*x*x)

# set the spacing between subplots
plt.subplots_adjust(left=0.1,
					bottom=0.1, 
					right=0.9, 
					top=0.9, 
					wspace=0.4, 
					hspace=0.4)
plt.show()
```
