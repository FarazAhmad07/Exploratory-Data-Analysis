### Box plot:
A box plot (or box-and-whisker plot) s is the visual representation of the depicting groups of numerical data through their quartiles against continuous/categorical data.

A box plot consists of 5 things.

    .Minimum
    .First Quartile or 25%
    .Median (Second Quartile) or 50%
    .Third Quartile or 75%
    .Maximum

In a box plot, the whiskers represent the spread of the data within 1.5 times the interquartile range (IQR) from the first quartile (Q1) and the third quartile (Q3).

The area inside the box (50% of the data) is known as the Inter Quartile Range (IQR), which is calculated as IQR = Q3 - Q1. Outliers are the data points below and above the lower and upper limit. The lower and upper limit is calculated as Lower Limit/Whisker = Q1 - 1.5*IQR and Upper Limit/Whisker = Q3 + 1.5*IQR. The values below and above these limits are considered outliers1.

    Box plots are useful for identifying outliers, visualizing the spread and skewness of the data, and comparing distributions

### Violin Plot:
A violin plot is similar to a boxplot. It shows several quantitative data across one or more categorical variables such that those distributions can be compared. 

#### Syntax: seaborn.violinplot(x=None, y=None, hue=None, data=None)

### Parameters: 

x, y, hue: Inputs for plotting long-form data. 
data: Dataset for plotting. 

#### A violin plot is a method to visualize the distribution of numerical data of different variables1. It combines a box plot and a kernel density plot to provide detailed information about the distribution.

Here’s a breakdown of its components:

    Density curves: These depict distributions of numeric data for one or more groups. The width of each curve corresponds with the approximate frequency of data points in each region.

    Box plot: Frequently overlaid on the density curves, it provides additional information such as the median (represented by a white dot), quartiles (the thin gray bar), and outliers.

    Kernel Density Estimation (KDE): This statistical technique is used to estimate the probability density function (PDF) of a random variable based on a set of observed data points.

#### Violin plots are especially useful when you want to make a comparison of distributions between multiple groups. The peaks, valleys, and tails of each group’s density curve can be compared to see where groups are similar or different.

### Bar plot:
Barplot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. 

Syntax : seaborn.barplot(x=None, y=None, hue=None, data=None)

Parameters :

x, y : This parameter take names of variables in data or vector data, Inputs for plotting long-form data.
hue : (optional) This parameter take column name for colour encoding.
data : (optional) This parameter take DataFrame, array, or list of arrays, Dataset for plotting. If x and y are absent, this is interpreted as wide-form. Otherwise it is expected to be long-form.
Returns : Returns the Axes object with the plot drawn onto it. 

```python
fig, ax = plt.subplots(figsize=(12, 6)) 
sns.set(style = 'whitegrid') 
 
# read csv and plot
#data = pandas.read_csv("nba.csv")
sns.barplot(x ="Age", y ="Weight", data = data, ax = ax) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure and axis with larger width
fig, ax = plt.subplots(figsize=(12, 6)) 

# Set the style
sns.set(style='whitegrid')

# Create the barplot
barplot = sns.barplot(x="Age", y="Weight", data=data, ax=ax)

# Calculate the total
total = len(data)

# Add annotations
for p in barplot.patches:
    height = p.get_height()
    barplot.text(p.get_x()+p.get_width()/2.,
            height + 13,
            '{:1.2f}'.format(height/total),
            ha="center") 

# Show the plot
plt.show()
```

    .for p in barplot.patches: This line starts a loop that goes through each bar (patch) in the bar plot.

    .height = p.get_height() This line gets the height of the current bar, which is the same as its y-value.

    .barplot.text(...) This function adds a text annotation to the bar plot.
    
    .p.get_x()+p.get_width()/2. This is the x-coordinate of the annotation. It’s set to the x-coordinate of the bar plus half its width, which places the annotation in the center of the bar.
    
    .height + 3 This is the y-coordinate of the annotation. It’s set to the height of the bar plus a small offset (3 in this case), which places the annotation just above the bar.
    
    .'{:1.2f}'.format(height/total) This is the text of the annotation. It’s set to the height of the bar divided by the total number of data points, formatted as a decimal with two digits after the decimal point. This gives the percentage that this bar represents out of the total.
    
    .ha="center" This argument sets the horizontal alignment of the annotation to center, which centers the annotation over the bar.

```python
import seaborn as sns
import pandas as pd

data = ["A", "A", "B", "B", "B", "C"]

# Create a DataFrame
df = pd.DataFrame(data, columns=['Variable'])

# Bar plot
sns.barplot(x = df['Variable'].value_counts().index, y = df['Variable'].value_counts().values)
```

In this code, we first convert the list into a DataFrame. Then, we use the value_counts() function to count the occurrences of each unique value in the data. The barplot() function then plots these counts. The x-values are the unique values from the data, and the y-values are their corresponding counts. This will give you a bar chart similar to the countplot.
